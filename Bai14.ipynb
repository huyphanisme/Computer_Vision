```python
import numpy as np
from PIL import Image

def average_images(image_list,size   ):
    total_array = np.array(Image.open(image_list[0]).resize(size), 'f')
    count=1

    for image_path in image_list[1:]:
        try:
            image_array = np.array(Image.open(image_path).resize(size), 'f')
            total_array += image_array
            count=count+1
        except Exception as e:
            print("Skip  {image_path}: {e}")

    # Tnh trung bnh
    average_array = total_array / count

    # Chuyn i kt qu trung bnh thnh hnh nh
    average_image = Image.fromarray(average_array.astype('uint8'))

    return average_image

# S dng v d:
my_path_1 = "D:/Computer_Vision/Computer_Vision/Images/Hinh6.jpg"
my_path_2 = "D:/Computer_Vision/Computer_Vision/Images/Hinh7.jpg"
image_list = [my_path_1, my_path_2]
size = (700, 500)
result_image = average_images(image_list, size)
display(result_image)


# S dng v d:
my_path_1 = "D:/Computer_Vision/Computer_Vision/Images/Hinh8.jpg"
my_path_2 = "D:/Computer_Vision/Computer_Vision/Images/Hinh9.jpg"
size = (700, 500)
result_image = average_images(image_list, size)
display(result_image)
```
